```python
import h5py
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
dataset_path = "/scr2/muj_senssearch_dataset10HZ_slide2020415/"

def loadread_dataset(filename):
    return h5py.File(filename, 'r', swmr=True, libver = 'latest')

def loadedit_dataset(filename):
    return h5py.File(filename, 'r+')
```

```python
filename_list = []
for file in os.listdir(dataset_path):
    if file.endswith(".h5"): # and len(filename_list) < 20:
        filename_list.append(dataset_path + file)

print(len(filename_list))
for filename in filename_list:
    print(filename)
    f = loadedit_dataset(filename)
#     print('insertion' in f.keys())
    hole_type = np.array(f['hole_type'])
    peg_type = np.array(f['peg_type'])
        
    height = np.array(f['proprio'])[:,2]
    width = np.array(f['proprio'])[:,0]
    length = np.array(f['proprio'])[:,1]
    insertion = np.zeros(height.shape[0])
    
    for i in range(height.shape[0]):
        h = height[i]
        w = width[i]
        l = length[i]

        if h < -0.002 and abs(w) < 0.0015\
        and abs(w) < 0.0015 and hole_type.argmax() == peg_type.argmax():
            insertion[i] = 1.0
        else:
            insertion[i] = 0.0

    if 'insertion' not in f.keys():
        print(filename)
        f.create_dataset("insertion", data = insertion)
        
    f.close()
# plt.plot(, dist_list)
# plt.show()

# print(contact_sum)
# print(total_points)
```
